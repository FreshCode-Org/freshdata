# freshdata quickstart

**Automated, explainable DataFrame cleaning for pandas.**

This notebook cleans a small messy dataset in one call, reads the audit
report, and previews the engine's decisions before applying them.

```bash
pip install freshdata
```

````python
import pandas as pd
import freshdata as fd

print('freshdata', fd.__version__)
````

## A deliberately messy frame

Whitespace, `N/A`/`-` sentinels, currency strings, an all-empty column,
a duplicate row, and everything stored as text.

````python
df = pd.DataFrame({
    'First Name': [' Ann ', 'Bob', 'Bob', 'Cara', None],
    'AGE': ['34', 'N/A', 'N/A', '41', '29'],
    'Joined Date': ['2021-01-01', '2021-02-15', '2021-02-15', '-', '2021-03-30'],
    'Active': ['yes', 'no', 'no', 'yes', 'no'],
    'Salary($)': ['$1,200.50', '-', '-', '$2,000', '$1,750'],
    'empty': [None, None, None, None, None],
})
df
````

## Profile before cleaning

`fd.profile` is read-only and uses the same inference as `fd.clean`.

````python
print(fd.profile(df))
````

## Clean with a full audit trail

````python
cleaned, report = fd.clean(df, return_report=True)
print(report.summary())
````

````python
cleaned
````

## The report is machine-readable too

````python
report.to_frame()
````
